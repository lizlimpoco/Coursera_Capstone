## Nearby venues per barangays

### Function that returns nearby venues

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Barangay', 
                  'Barangay Latitude', 
                  'Barangay Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### Frequency of venue categories

```python
# one hot encoding
dvobrgys_onehot = pd.get_dummies(dvobrgys_venues[['Venue Category']], prefix="", prefix_sep="")

# add barangay column back to dataframe
dvobrgys_onehot['Barangay'] = dvobrgys_venues['Barangay']

# move barangay column to the first column
fixed_columns = [dvobrgys_onehot.columns[-1]] + list(dvobrgys_onehot.columns[:-1])
dvobrgys_onehot = dvobrgys_onehot[fixed_columns]

dvobrgys_onehot.head()
```

```python
dvobrgys_vcfreq = dvobrgys_onehot.groupby('Barangay').mean().reset_index().round(2)
dvobrgys_vcfreq.head()
```

## k-Means Clustering

```python
to_cluster = dvobrgys_vcfreq
for i in range(to_cluster.shape[0]):
    for j in range(dvobrgys_df.shape[0]):
        if(to_cluster.loc[i,'Barangay']==dvobrgys_df.loc[j,'Barangay']):
            to_cluster.loc[i,'Latitude'] = dvobrgys_df.loc[j,'Latitude']
            to_cluster.loc[i,'Longitude'] = dvobrgys_df.loc[j,'Longitude']
to_cluster.head()
```

```python
from sklearn.cluster import KMeans

# set number of clusters
kclusters = 3

dvobrgys_vcfreq_clustering = to_cluster.drop(['Barangay'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dvobrgys_vcfreq_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

## Sort venue categories

```python
dvobrgys_vcfreq.drop(['Latitude','Longitude'],axis=1,inplace=True)
dvobrgys_vcfreq.head()
```

### Function

```python
def return_sorted_venues(row, num_top_venues):
    row_categories_sorted = row.sort_values(ascending=False)
    for i in range(len(row_categories_sorted)):
        if(row_categories_sorted.values[i]==0):
            row_categories_sorted.values[i] = '---'
        else:
            row_categories_sorted.values[i] = row_categories_sorted.index.values[i]+", "+str(row_categories_sorted.values[i])
    return row_categories_sorted.values[0:num_top_venues]
```

### Show top 20 venue categories per barangay

```python
import numpy as np

num_top_venues = 20

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Barangay']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
dvobrgys_venues_sorted = pd.DataFrame(columns=columns)
dvobrgys_venues_sorted['Barangay'] = dvobrgys_vcfreq['Barangay']

for ind in np.arange(dvobrgys_vcfreq.shape[0]):
    dvobrgys_venues_sorted.iloc[ind, 1:] = return_sorted_venues(dvobrgys_vcfreq.iloc[ind, 1:], num_top_venues)
    

dvobrgys_venues_sorted.head()
```
